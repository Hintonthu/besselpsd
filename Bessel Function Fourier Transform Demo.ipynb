```python
import numpy as np
import scipy.signal as ss
import scipy.special as sp
import matplotlib.pyplot as plt
```

```python
def getPSD(fs, tmax):
    numpoints = tmax * fs
    t = np.linspace(0, tmax, num=numpoints)
    x = sp.jv(0, t)
    f, Pxx = ss.periodogram(x, fs)
    Pxx = Pxx * (1.0 / Pxx[1])
    return f, Pxx

def plotPSD(fs, tmax):
    f, Pxx = getPSD(fs, tmax)
    plt.semilogx(f, 10*np.log10(Pxx))
    plt.ylim(-100, 20)
    plt.xlim(1e-2, 1e2)
    plt.grid()
    plt.show()
```

```python
plotPSD(1e3, 1000)
```

```python
plotPSD(1e5, 100)
```

```python
plotPSD(1e3, 10000)
```
